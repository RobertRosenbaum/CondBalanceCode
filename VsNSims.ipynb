```python
import numpy as np
from time import time as tm

# import stuff from our own creations
from default_params import *
from utils import *
from NetworkModels import *

np.random.seed(1)

runsims=True
```

```python

# Duration of simulation in ms
T = 5000
dt = 0.1
time = np.arange(0,T,dt)
Nt = len(time)

# Burn-in period for compring time averages and rates
Tburn = 250

dtRecord = 1
timeRecord = np.arange(0,T,dtRecord)
NtRecord = len(timeRecord)
Nburn = int(Tburn/dtRecord)

# Disctretized time
time = np.arange(0,T,dt)
Nt = len(time)
Nburn = int(Tburn/dt)


## run sims for different N and c for current-based andconductance-based synapses
Ns = np.array([1000, 2000, 4000, 8000, 15000])
cs = np.array([0,0.2])
numtrials = 4

# File name for saving data
filename = './Data/VsNPlotData.npy'

```

```python


if runsims:
  # Run sim
  storeVarVeCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeVarVeEff = np.zeros((numtrials, len(Ns), len(cs)))
  storeVarVe = np.zeros((numtrials, len(Ns), len(cs)))

  storeMeanVeCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeMeanVeEff = np.zeros((numtrials, len(Ns), len(cs)))
  storeMeanVe = np.zeros((numtrials, len(Ns), len(cs)))

  # # These arrays need to be initialized inside the loop over Ns since Ne changes
  # storeAllMeanVe = []
  # storeAllVarVe = []
  # storeAllMeanVeCond = []
  # storeAllVarVeCond = []
  # storeAllMeanVeEff = []
  # storeAllVarVeEff = []

  # storeAllMeanVe = np.zeros((numtrials, len(Ns), len(cs), Ne))
  # storeAllVarVe = np.zeros((numtrials, len(Ns), len(cs), Ne))
  # storeAllMeanVeCond = np.zeros((numtrials, len(Ns), len(cs), Ne))
  # storeAllVarVeCond = np.zeros((numtrials, len(Ns), len(cs), Ne))
  # storeAllMeanVeEff = np.zeros((numtrials, len(Ns), len(cs), Ne))
  # storeAllVarVeEff = np.zeros((numtrials, len(Ns), len(cs), Ne))



  # storeVarVeCondFree = np.zeros((numtrials, len(Ns), len(cs)))
  # storeVarVeEffFree = np.zeros((numtrials, len(Ns), len(cs)))
  # storeVarVeFree = np.zeros((numtrials, len(Ns), len(cs)))

  # storeMeanVeCondFree = np.zeros((numtrials, len(Ns), len(cs)))
  # storeMeanVeEffFree = np.zeros((numtrials, len(Ns), len(cs)))
  # storeMeanVeFree = np.zeros((numtrials, len(Ns), len(cs)))

  StoreCorrSS = np.zeros((numtrials, len(Ns), len(cs)))
  StoreCorrSSCond = np.zeros((numtrials, len(Ns), len(cs)))
  StoreCorrSSEff = np.zeros((numtrials, len(Ns), len(cs)))

  StoreShuffledCorrSS = np.zeros((numtrials, len(Ns), len(cs)))
  StoreShuffledCorrSSCond = np.zeros((numtrials, len(Ns), len(cs)))
  StoreShuffledCorrSSEff = np.zeros((numtrials, len(Ns), len(cs)))


  # Number of bins for histograms
  numbins = 200

  storeHistVeCond = np.zeros((numtrials, len(Ns), len(cs), numbins))
  storeBinsVeCond = np.zeros((numtrials, len(Ns), len(cs), numbins+1))
  storeHistVeEff = np.zeros((numtrials, len(Ns), len(cs), numbins))
  storeBinsVeEff = np.zeros((numtrials, len(Ns), len(cs), numbins+1))
  storeHistVe = np.zeros((numtrials, len(Ns), len(cs), numbins))
  storeBinsVe = np.zeros((numtrials, len(Ns), len(cs), numbins+1))

  storeSeCond = [[None]*len(cs)]*len(Ns)
  storeSiCond = [[None]*len(cs)]*len(Ns)
  storeSeEff = [[None]*len(cs)]*len(Ns)
  storeSiEff = [[None]*len(cs)]*len(Ns)
  storeSe = [[None]*len(cs)]*len(Ns)
  storeSi = [[None]*len(cs)]*len(Ns)

  storeReCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeRiCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeReEff = np.zeros((numtrials, len(Ns), len(cs)))
  storeRiEff = np.zeros((numtrials, len(Ns), len(cs)))
  storeRe = np.zeros((numtrials, len(Ns), len(cs)))
  storeRi = np.zeros((numtrials, len(Ns), len(cs)))

  storeVarIexCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeVarIexEff = np.zeros((numtrials, len(Ns), len(cs)))
  storeVarIex = np.zeros((numtrials, len(Ns), len(cs)))

  storeErrorTerm = np.zeros((numtrials, len(Ns), len(cs)))
  storeAbsErrorTerm = np.zeros((numtrials, len(Ns), len(cs)))
  storeDeeCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeDeiCond = np.zeros((numtrials, len(Ns), len(cs)))
  #storeDieCond = np.zeros((numtrials, len(Ns), len(cs)))
  #storeDiiCond = np.zeros((numtrials, len(Ns), len(cs)))
  storeDexCond = np.zeros((numtrials, len(Ns), len(cs)))
  #storeDixCond = np.zeros((numtrials, len(Ns), len(cs)))

  storegLEff = np.zeros((numtrials, len(Ns), len(cs), 2))


  Ee = NeuronParamsCond['Ee']
  Ei = NeuronParamsCond['Ei']
  EL = NeuronParamsCond['EL']

  gL = NeuronParamsCond['gL']
  Cm = NeuronParamsCond['Cm']



  t00=tm()
  for iii in range(numtrials):
    print('Starting trial',iii+1,'of',numtrials,'after',tm()-t00,'s')
    for n in range(len(Ns)):
        t0N=tm()
        N = Ns[n]
        Ne = int(N*.8)
        Ni = N-Ne
        Nx = int(N*qx)

        JmCond = np.array([[jeeCond, jeiCond], [jieCond, jiiCond]])/np.sqrt(N)
        JxmCond = np.array([[jexCond], [jixCond]])/np.sqrt(N)

        Jm = np.array([[jee, jei], [jie, jii]])/np.sqrt(N)
        Jxm = np.array([[jex], [jix]])/np.sqrt(N)


        # Max number of spikes
        maxns=int(N*(30/1000)*T)

        # Which neurons to record from
        IeRecord=np.arange(Ne)
        IiRecord=np.arange(Ni)


        for c in range(len(cs)):

          #########
          #### Conductance-based sim
          #########
          # Really, we should be able to do this outside of the c loop
          # but this is better because we can delete the JCond's before
          # generating the J's to reduce total mem usage
          t0=tm()
          #print('starting cond J',N)
          JCond = GetBlockErdosRenyi((Ne,Ni),JmCond,P)
          JxCond = GetBlockErdosRenyi((Nx,),JxmCond,Px,(Ne,Ni))
          tJ = tm()-t0
          #print('Time for cond J:',tJ,'s',N)

          # Set params and variables that depend on c (just Sx)
          t0=tm()
          #print('starting Sx',N)
          Sx = PoissonProcess(rx,Nt,dt,Nx,cs[c],rep='sparse',taujitter=taujitter)
          tSx = tm()-t0
          #print('Time for cond Sx:',tSx,'s')


          t0 = tm()
          #print('starting cond sim',N)
          se,si,Recording=EIFNetworkCond(JCond, Sx, JxCond, Ne, NeuronParamsCond, tau, Nt, dt, maxns, IeRecord, IiRecord, dtRecord)
          tSim2=tm()-t0
          print('Time for cond sim:',tSim2,'s')


          Ve = Recording['Ve'][Nburn:,:]
          Vi = Recording['Vi'][Nburn:,:]
          #VeFree = Recording['VeFree'][Nburn:,:]
          gee = Recording['gee'][Nburn:,:]
          gei = Recording['gei'][Nburn:,:]
          gie = Recording['gie'][Nburn:,:]
          gii = Recording['gii'][Nburn:,:]
          gex = Recording['gex'][Nburn:,:]
          gix = Recording['gix'][Nburn:,:]
          meanVe = Ve.mean()
          meanVi = Vi.mean()
          varVe = Ve.var(axis=0).mean()          
          varV = np.concatenate([Ve, Vi], axis=1).var(axis=0).mean()

          #varVeFree = VeFree.var(axis=0).mean()
          #meanVeFree = VeFree.mean()
          varIex = (gex*(Ve-Ee)).var(axis=0).mean()


          storeVarVeCond[iii,n,c] = varV
          #storeVarVeCondFree[iii,n,c] = varVeFree
          storeMeanVeCond[iii,n,c] = meanVe
          #storeMeanVeCondFree[iii,n,c] = meanVeFree
          h,b = np.histogram(Ve, bins=numbins, density=True)
          storeHistVeCond[iii,n,c,:] = h
          storeBinsVeCond[iii,n,c,:] = b
          storeVarIexCond[iii,n,c] = varIex

          # storeAllMeanVeCond[iii,n,c,:] = Ve.mean(axis=0)
          # storeAllVarVeCond[iii,n,c,:] = Ve.var(axis=0)

          #storeDeeCond[iii,n,c] = np.mean((Ve - Ve.mean(axis = 0))*gee*(Ve-Ee))
          V = np.concatenate([Ve, Vi], axis=1)
          ge = np.concatenate([gee, gie], axis=1)
          gi = np.concatenate([gei, gii], axis=1)
          gx = np.concatenate([gex, gix], axis=1)
          geff = ge+gi+gx+NeuronParams['gL']
          storeErrorTerm[iii,n,c] = ((V-V.mean(axis = 0))*(geff-geff.mean(axis = 0))).mean()
          storeAbsErrorTerm[iii,n,c] = np.abs(((V-V.mean(axis = 0))*(geff-geff.mean(axis = 0)))).mean()
          storeDeeCond[iii,n,c] = np.mean((Ve - Ve.mean(axis = 0))*(gee-gee.mean(axis = 0)))
          storeDeiCond[iii,n,c] = np.mean((Ve - Ve.mean(axis = 0))*(gei-gei.mean(axis = 0)))
          storeDexCond[iii,n,c] = np.mean((Ve - Ve.mean(axis = 0))*(gex-gex.mean(axis = 0)))
          storeReCond[iii,n,c] = np.count_nonzero(se[0,:]>Tburn)/((T-Tburn)*Ne)
          storeRiCond[iii,n,c] = np.count_nonzero(si[0,:]>Tburn)/((T-Tburn)*Ni)
          if iii==0:
            storeSeCond[n][c] = se.copy()
            storeSiCond[n][c] = si.copy()
          ### Get all spikes and sort them
          s = np.concatenate((se, si), axis=1)
          sorted_indices = np.argsort(s[0, :])
          s = s[:, sorted_indices]
          ###
          StoreCorrSSCond[iii,n,c] = GetSpikeCountCorrelation(s, N, T, winsize, min_rate=1/1000).mean()
          StoreShuffledCorrSSCond[iii,n,c] = GetShuffledSpikeCountCorrelation(s, N, T, winsize).mean()


          #########
          #### Current-based sim using effective leak and weights
          #########
          jeeEff = jeeCond*(Ee-meanVe)
          jeiEff = jeiCond*(Ei-meanVe)
          jieEff = jieCond*(Ee-meanVi)
          jiiEff = jiiCond*(Ei-meanVi)
          jexEff = jexCond*(Ee-meanVe)
          jixEff = jixCond*(Ee-meanVi)

          JmEff = np.array([[jeeEff, jeiEff], [jieEff, jiiEff]])/np.sqrt(N)
          JxmEff = np.array([[jexEff], [jixEff]])/np.sqrt(N)

          

          NeuronParamsEff = NeuronParamsCond.copy()
          del NeuronParamsEff['Ee']
          del NeuronParamsEff['Ei']
          gLeff = np.zeros(2)
          gLeff[0] = gL+gee.mean()+gei.mean()+gex.mean()
          gLeff[1] = gL+gie.mean()+gii.mean()+gix.mean()
          storegLEff[iii,n,c,:] = gLeff
          ELeff = np.zeros(2)
          ELeff[0] = meanVe
          ELeff[1] = meanVi
          NeuronParamsEff['gL'] = gLeff
          NeuronParamsEff['EL'] = ELeff



          del JCond, JxCond, Recording, se, si
          del Ve,Vi,gee,gei,gie,gii,gex,gix


          # Get connectivity matrix
          t0=tm()
          JEff = GetBlockErdosRenyi((Ne,Ni),JmEff,P)
          JxEff = GetBlockErdosRenyi((Nx,),JxmEff,Px,(Ne,Ni))
          tJ = tm()-t0
          ###


          # Run sim with effective leak
          t0 = tm()
          se,si,Recording=EIFNetworkCurrent(JEff, Sx, JxEff, Ne, NeuronParamsEff, tau, Nt, dt, maxns, IeRecord, IiRecord, dtRecord)
          tSim2=tm()-t0
          print('Time for curr sim Eff:',tSim2,'s')

          Ve = Recording['Ve'][Nburn:,:]
          Vi = Recording['Vi'][Nburn:,:]
          #VeFree = Recording['VeFree'][Nburn:,:]
          # Iee = Recording['Iee'][Nburn:,:]
          # Iei = Recording['Iei'][Nburn:,:]
          # Iie = Recording['Iie'][Nburn:,:]
          # Iii = Recording['Iii'][Nburn:,:]
          Iex = Recording['Iex'][Nburn:,:]
          #Iix = Recording['Iix'][Nburn:,:]
          meanVe = Ve.mean()
          meanVi = Vi.mean()
          #meanVeFree = VeFree.mean()
          varVe = Ve.var(axis=0).mean()
          #varVeFree = VeFree.var(axis=0).mean()
          varIex = Iex.var(axis=0).mean()
          varV = np.concatenate([Ve, Vi], axis=1).var(axis=0).mean()

          storeVarVeEff[iii,n,c] = varV
          #storeVarVeEffFree[iii,n,c] = varVeFree
          storeMeanVeEff[iii,n,c] = meanVe
          #storeMeanVeEffFree[iii,n,c] = meanVeFree
          h,b = np.histogram(Ve, bins=numbins, density=True)
          storeHistVeEff[iii,n,c,:] = h
          storeBinsVeEff[iii,n,c,:] = b
          storeVarIexEff[iii,n,c] = varIex


          # storeAllMeanVeEff[iii,n,c,:] = Ve.mean(axis=0)
          # storeAllVarVeEff[iii,n,c,:] = Ve.var(axis=0)

          storeReEff[iii,n,c] = np.count_nonzero(se[0,:]>Tburn)/((T-Tburn)*Ne)
          storeRiEff[iii,n,c] = np.count_nonzero(si[0,:]>Tburn)/((T-Tburn)*Ni)
          if iii==0:
            storeSeEff[n][c] = se.copy()
            storeSiEff[n][c] = si.copy()
          ### Get all spikes and sort them
          s = np.concatenate((se, si), axis=1)
          sorted_indices = np.argsort(s[0, :])
          s = s[:, sorted_indices]
          ###
          StoreCorrSSEff[iii,n,c] = GetSpikeCountCorrelation(s, N, T, winsize, min_rate=1/1000).mean()
          StoreShuffledCorrSSEff[iii,n,c] = GetShuffledSpikeCountCorrelation(s, N, T, winsize).mean()

          del Recording, se, si
          del Ve,Vi,Iex#,VeFree#Iee,Iei,Iie,Iii,Iix


          #########
          #### Naive current-based sim
          #########
          del JEff, JxEff
          J = GetBlockErdosRenyi((Ne,Ni),Jm,P)
          Jx = GetBlockErdosRenyi((Nx,),Jxm,Px,(Ne,Ni))

          NeuronParams = NeuronParamsCond.copy()
          del NeuronParams['Ee']
          del NeuronParams['Ei']


          t0 = tm()
          se,si,Recording=EIFNetworkCurrent(J, Sx, Jx, Ne, NeuronParams, tau, Nt, dt, maxns, IeRecord, IiRecord, dtRecord)
          tSim2=tm()-t0
          print('Time for curr sim:',tSim2,'s')

          Ve = Recording['Ve'][Nburn:,:]
          Vi = Recording['Vi'][Nburn:,:]
          #VeFree = Recording['VeFree'][Nburn:,:]
          #Iee = Recording['Iee'][Nburn:,:]
          #Iei = Recording['Iei'][Nburn:,:]
          #Iie = Recording['Iie'][Nburn:,:]
          #Iii = Recording['Iii'][Nburn:,:]
          Iex = Recording['Iex'][Nburn:,:]
          #Iix = Recording['Iix'][Nburn:,:]
          meanVe = Ve.mean()
          meanVi = Vi.mean()
          #meanVeFree = VeFree.mean()
          varVe = Ve.var(axis=0).mean()
          #varVeFree = VeFree.var(axis=0).mean()
          varIex = Iex.var(axis=0).mean()
          varV = np.concatenate([Ve, Vi], axis=1).var(axis=0).mean()

          storeVarVe[iii,n,c] = varV
          #storeVarVeFree[iii,n,c] = varVeFree
          storeMeanVe[iii,n,c] = meanVe
          #storeMeanVeFree[iii,n,c] = meanVeFree

          # storeAllMeanVe[iii,n,c,:] = Ve.mean(axis=0)
          # storeAllVarVe[iii,n,c,:] = Ve.var(axis=0)


          h,b = np.histogram(Ve, bins=numbins, density=True)
          storeHistVe[iii,n,c,:] = h
          storeBinsVe[iii,n,c,:] = b
          storeVarIex[iii,n,c] = varIex
          storeRe[iii,n,c] = np.count_nonzero(se[0,:]>Tburn)/((T-Tburn)*Ne)
          storeRi[iii,n,c] = np.count_nonzero(si[0,:]>Tburn)/((T-Tburn)*Ni)
          if iii==0:
            storeSe[n][c] = se.copy()
            storeSi[n][c] = si.copy()
          ### Get all spikes and sort them
          s = np.concatenate((se, si), axis=1)
          sorted_indices = np.argsort(s[0, :])
          s = s[:, sorted_indices]
          ###
          StoreCorrSS[iii,n,c] = GetSpikeCountCorrelation(s, N, T, winsize, min_rate=1/1000).mean()
          StoreShuffledCorrSS[iii,n,c] = GetShuffledSpikeCountCorrelation(s, N, T, winsize).mean()

          del Recording, se, si
          del Ve,Vi,Iex#Iee,Iei,Iie,Iii,Iix
          del J, Jx
        print('*** N=',N,';',n+1,'of',len(Ns),'done in',tm()-t0N,'s on trial',iii+1,'of',numtrials,'***\n')

  totaltime = (tm()-t00)
  print('Total time:',totaltime/60.0,'min')


  ########################################################### save data ###########################################################

  data = {}
  data['Ns']=Ns
  data['cs']=cs
  data['T']=T
  data['storeVarVeCond']=storeVarVeCond
  data['storeVarVeEff']=storeVarVeEff
  data['storeVarVe']=storeVarVe
  #data['storeVarVeCondFree']=storeVarVeCondFree
  #data['storeVarVeEffFree']=storeVarVeEffFree
  #data['storeVarVeFree']=storeVarVeFree
  data['storeMeanVeCond']=storeMeanVeCond
  data['storeMeanVeEff']=storeMeanVeEff
  data['storeMeanVe']=storeMeanVe
  #data['storeMeanVeCondFree']=storeMeanVeCondFree
  #data['storeMeanVeEffFree']=storeMeanVeEffFree
  #data['storeMeanVeFree']=storeMeanVeFree

  # data['storeAllMeanVeCond']=storeAllMeanVeCond
  # data['storeAllVarVeCond']=storeAllVarVeCond
  # data['storeAllMeanVeEff']=storeAllMeanVeEff
  # data['storeAllVarVeEff']=storeAllVarVeEff
  # data['storeAllMeanVe']=storeAllMeanVe
  # data['storeAllVarVe']=storeAllVarVe

  data['numbins']=numbins
  data['storeHistVeCond']=storeHistVeCond
  data['storeBinsVeCond']=storeBinsVeCond
  data['storeHistVeEff']=storeHistVeEff
  data['storeBinsVeEff']=storeBinsVeEff
  data['storeHistVe']=storeHistVe
  data['storeBinsVe']=storeBinsVe
  data['storeSeCond']=storeSeCond
  data['storeSiCond']=storeSiCond
  data['storeSeEff']=storeSeEff
  data['storeSiEff']=storeSiEff
  data['storeSe']=storeSe
  data['storeSi']=storeSi
  data['storeReCond']=storeReCond
  data['storeRiCond']=storeRiCond

  data['StoreCorrSS']=StoreCorrSS
  data['StoreCorrSSCond']=StoreCorrSSCond
  data['StoreCorrSSEff']=StoreCorrSSEff

  data['StoreShuffledCorrSS']=StoreShuffledCorrSS
  data['StoreShuffledCorrSSCond']=StoreShuffledCorrSSCond
  data['StoreShuffledCorrSSEff']=StoreShuffledCorrSSEff

  data['storeReEff']=storeReEff
  data['storeRiEff']=storeRiEff
  data['storeRe']=storeRe
  data['storeRi']=storeRi
  data['storeVarIexCond']=storeVarIexCond
  data['storeVarIexEff']=storeVarIexEff
  data['storeVarIex']=storeVarIex
  data['storeErrorTerm']=storeErrorTerm
  data['storeAbsErrorTerm']=storeAbsErrorTerm

  data['storeDeeCond']=storeDeeCond
  data['storeDeiCond']=storeDeiCond
  #data['storeDieCond']=storeDieCond
  #data['storeDiiCond']=storeDiiCond
  data['storeDexCond']=storeDexCond
  #data['storeDixCond']=storeDixCond
  data['storegLEff']=storegLEff

  data['time']=time
  data['timeRecord']=timeRecord
  data['numtrials']=numtrials
  data['NeuronParamsCond']=NeuronParamsCond
  data['JmCond']=JmCond
  data['JxmCond']=JxmCond

  data['rBal']=rBal

  np.save(filename, data)


else:
  ########################################################### load data ###########################################################
  saved_data = np.load(filename, allow_pickle=True).item()
  Ns = saved_data['Ns']
  cs = saved_data['cs']
  T = saved_data['T']
  storeVarVeCond = saved_data['storeVarVeCond']
  storeVarVeEff = saved_data['storeVarVeEff']
  storeVarVe = saved_data['storeVarVe']
  #storeVarVeCondFree = saved_data['storeVarVeCondFree']
  #storeVarVeEffFree = saved_data['storeVarVeEffFree']
  #storeVarVeFree = saved_data['storeVarVeFree']
  storeMeanVeCond = saved_data['storeMeanVeCond']
  storeMeanVeEff = saved_data['storeMeanVeEff']
  storeMeanVe = saved_data['storeMeanVe']

  #storeMeanVeCondFree = saved_data['storeMeanVeCondFree']
  #storeMeanVeEffFree = saved_data['storeMeanVeEffFree']
  #storeMeanVeFree = saved_data['storeMeanVeFree']

  # storeAllMeanVeCond = saved_data['storeAllMeanVeCond']
  # storeAllVarVeCond = saved_data['storeAllVarVeCond']
  # storeAllMeanVeEff = saved_data['storeAllMeanVeEff']
  # storeAllVarVeEff = saved_data['storeAllVarVeEff']
  # storeAllMeanVe = saved_data['storeAllMeanVe']
  # storeAllVarVe = saved_data['storeAllVarVe']
  
  numbins = saved_data['numbins']
  storeHistVeCond = saved_data['storeHistVeCond']
  storeBinsVeCond = saved_data['storeBinsVeCond']
  storeHistVeEff = saved_data['storeHistVeEff']
  storeBinsVeEff = saved_data['storeBinsVeEff']
  storeHistVe = saved_data['storeHistVe']
  storeBinsVe = saved_data['storeBinsVe']
  storeSeCond = saved_data['storeSeCond']
  storeSiCond = saved_data['storeSiCond']
  storeSeEff = saved_data['storeSeEff']
  storeSiEff = saved_data['storeSiEff']
  storeSe = saved_data['storeSe']
  storeSi = saved_data['storeSi']
  storeReCond = saved_data['storeReCond']
  storeRiCond = saved_data['storeRiCond']

  StoreCorrSS = saved_data['StoreCorrSS']
  StoreCorrSSCond = saved_data['StoreCorrSSCond']
  StoreCorrSSEff = saved_data['StoreCorrSSEff']

  storeReEff = saved_data['storeReEff']
  storeRiEff = saved_data['storeRiEff']
  storeRe = saved_data['storeRe']
  storeRi = saved_data['storeRi']
  storeVarIexCond = saved_data['storeVarIexCond']
  storeVarIexEff = saved_data['storeVarIexEff']
  storeVarIex = saved_data['storeVarIex']
  storeErrorTerm = saved_data['storeErrorTerm']
  storeDeeCond = saved_data['storeDeeCond']
  storeDeiCond = saved_data['storeDeiCond']
  #storeDieCond = saved_data['storeDieCond']
  #storeDiiCond = saved_data['storeDiiCond']
  storeDexCond = saved_data['storeDexCond']
  #storeDixCond = saved_data['storeDixCond']
  storegLEff = saved_data['storegLEff']

  time = saved_data['time']
  timeRecord = saved_data['timeRecord']
  numtrials = saved_data['numtrials']
  NeuronParamsCond = saved_data['NeuronParamsCond']

  rBal = saved_data['rBal']
  #JmCond = saved_data['JmCond']
  #JxmCond = saved_data['JxmCond']
```
